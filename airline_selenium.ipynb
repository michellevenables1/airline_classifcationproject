```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.select import Select
from bs4 import BeautifulSoup
import re
import pandas as pd
import os
import zipfile
```

```python
#airline
my_url = 'https://www.transtats.bts.gov/OT_Delay/OT_DelayCause1.asp?pn=1'

driver = webdriver.Chrome("./chromedriver")
driver.get(my_url)
```

```python
carrier_dropdown = Select(driver.find_element_by_id('Carrier'))
carrier_dropdown.select_by_index(0)

carrier_dropdown = Select(driver.find_element_by_id('Airport'))
carrier_dropdown.select_by_index(0)

carrier_dropdown = Select(driver.find_element_by_id('PeriodFrom'))
carrier_dropdown.select_by_value('24042') #from Aug2018 to Aug2019
```

```python
submit_button = driver.find_elements_by_xpath('//*[@id="submit"]')[0]
submit_button.click()
```

```python
file = driver.find_element_by_xpath('//*[@id="content"]/table/tbody/tr[2]/td/table/tbody/tr[2]/td[2]/table[1]/tbody/tr[9]/td[3]/a[2]')
```
